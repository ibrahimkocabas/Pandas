# Birleştirme (Join) İşlemleri

```python
import numpy as np
import pandas as pd
m = np.random.randint(1,30, size = (5,3))
df1 = pd.DataFrame(m,columns = ["var1","var2","var3"])
df1
```

```python
df2 = df1 + 99
```

```python
df2
```

```python
pd.concat([df1,df2])
```

Birleştirme işleminde görüleceği üzere index sıralamasında DataFrame'leri ayrı olarak verdi.
Bu index sıralamasını biçimlendirmek yani 0'dan başlayarak sıralı hale getirmek için **ignore_index = True** ifadesini ekleme yapıyoruz.

```python
pd.concat([df1,df2], ignore_index=True)
```

```python
df1.columns
```

```python
df2.columns
```

```python
df2.columns = ["var1","var2","deg3"]
```

```python
df2
```

```python
df1
```

```python
pd.concat([df1,df2])
```

**DataFrame'lerde birbirleriyle karşılığı olmayan değişikkenlerde birleştirme işlemi yapıldığı zaman görüldüğü üzere hata ile karşılaşıyoruz.**

```python
pd.concat([df1,df2], join="inner")
```

**Inner Join ile birleştirme yapıldığı zaman farklı olan iki değişkeninde kesisim kısmındaki değerleri
alarak bir birleştirme işlemi sağlıyor.**

```python
pd.concat([df1,df2], join_axes = [df1.columns],ignore_index = True)
```

```python
pd.concat([df1,df2.reindex(columns=df1.columns)], ignore_index = True)
```

Farklı değişken isimleri olsa dahi birleştirme işlemine devam etmek istediğimizde ;
Değer tutan değişken ismini **join_axes** ya da **reindex** içeresinde tanımlayarak yapabiliriz.
